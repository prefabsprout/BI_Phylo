## 1. Лесопосадки

### a) NJ

```python
with open('SUP35_NJ.nwk') as tree:
    nwk = [line for line in tree]
    tree = Phylo.read(StringIO(*nwk), 'newick')
Phylo.draw(tree)
```

### б) UPGMA

```python
with open('SUP35_UPGMA.nwk') as tree:
    nwk = [line for line in tree]
    tree = Phylo.read(StringIO(*nwk), 'newick')
Phylo.draw(tree)
```
